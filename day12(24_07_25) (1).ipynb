# matplotlib :matplotlib.pyplot is a plotting library used for 2D graphics in python programming language.
# It can be used in python scripts, shell, web application servers and other graphical user interface toolkits.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=([10,20,30,40,50])
y=([11,22,33,44,55])
plt.xlim(10,50)  # x and y value cab be fixed by the user
plt.ylim(11,55)
plt.title("default x and y")
plt.plot(x,y)
plt.show()
```

# functional method

```python
x=np.linspace(1,2,10) #1 can be start and ended to be 5,  splitted by 10
x
```

```python
x=np.linspace(1,2,10)
y=x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

# title and labels


```python
plt.xlabel("x values")     # label
plt.ylabel("y values")
plt.title("graph")         # title
plt.plot(x,y)
plt.show()
```

```python
x1=np.linspace(-np.pi,np.pi,200)
x2 = np.linspace(-np.pi, np.pi, 200)

y1 = np.cos(x1)
y2 = np.sin(x2)
plt.plot(x1,y1,x2,y2)
plt.show()
```

#figure size and dpi

```python
plt.fire(figsize=(11,5),dpi=100)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```
