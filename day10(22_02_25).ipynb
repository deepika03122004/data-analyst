# group by

```python
import pandas as pd
import numpy as np
data = {
    'Company':['Google', 'Google', 'MSFT', 'FB','FB','IBM'],
    'Person': ['Sam','Nihad','Any','Van','Rakib','Ovi'],
    'Sales': [200, 120, 340, 124, 243, 350]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
byComp=df.groupby('Company')
df.mean(numeric_only=True)
```

```python
byComp.sum()
```

```python
df
```

```python
df['Sales'].std()
```

```python
df.groupby('Company').count()
```

```python
df
```

```python
df.groupby('Company').describe()
```

# merging

```python
left = {
    'key':['k0','k1','k2','k3'],
    'A':[10,20,30,40],
    'B':[1,2,3,4]
}
```

```python
right = {
    'key':['k0','k1','k2','k3'],
    'C':[100,200,300,400],
    'D':[11,12,13,14]
}
```

```python
left = pd.DataFrame(left)
```

```python
right = pd.DataFrame(right)
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge
```

# joining

```python
left = {
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
}
```

```python
right = {
    'C':['C0','C2','C3'],
    'D':['D0','D2','D3']
}
```

```python
left = pd.DataFrame(left, index = ['k0','k1','k2'])
```

```python
right = pd.DataFrame(right, index = ['k0','k2','k3'])
```

```python
left
```

```python
right
```

```python
left.join(right, how = 'inner')
```

```python
left.join
```
