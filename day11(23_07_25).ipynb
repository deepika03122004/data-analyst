```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
z=[11,22,33,44,55]
plt.plot(x,y,color="red",linewidth=2,linestyle="--")
plt.title("styled line plot")
plt.grid(True)
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```
